```python
#En <df> crea la columna "FEchas".
#Haciendo uso de Pandas <pd>, usa la función <.to_datetime> para que a la columna
#  agarre los datos de la columna 'FECHA' y los meta en "FECHAS" con el formato de día/mes/año
#  <%d/%m/%Y>

df["FECHAS"] = pd.to_datetime(df['FECHA'], format='%d/%m/%Y')

#De <df>, usa la funci+on <.drop> para eliminar la columna 'FECHA'
df = df.drop('FECHA', 1)
df
```

```python
df.dtypes
```

```python
#Con la función <.groupby> y la librería datetime <dt>
#Selecciona los meses '%B" y calcula el promedio de las precipitaciones
#   ordenados de menor a mayor

df.groupby(df['FECHAS'].dt.strftime('%B'))['PRECIP'].mean().sort_values()

```

```python
df.groupby(df['FECHAS'].dt.strftime('%B'))['TMAX'].mean().sort_values()
```

```python
df.groupby(df['FECHAS'].dt.strftime('%B'))['TMIN'].mean().sort_values()
```

```python
df.groupby(df['FECHAS'].dt.strftime('%Y'))['PRECIP'].mean().sort_values()
```
